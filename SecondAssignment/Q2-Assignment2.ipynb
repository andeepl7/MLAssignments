The calculations follow the next function due to the layer is fully connected:
$$
  h_{W,b}(X) = \phi (XW+b)
$$

X represent the matrix of input features
W matrix contains all the connections weights
b bias vector
$\phi$ Activation function

Network calculations
1. Compute the output of the hidden layer and the output-layer neurons of the given output (0.5,1)

```python
#Parameters
X1 = [0.5, 1]
Hidden_Layer_Weights =  [[-1, 0],
                         [2, 1],
                         [-2,1]]

Output_Layer_Weights = [[2, -0.5, 1],
                        [-2, 1, 0.5]]

#Bias is assumed as 0
```

```python
X1 = np.array(X1)
HLW = np.array(Hidden_Layer_Weights)
OLW = np.array(Output_Layer_Weights)
```

```python
#Output of the hidden layer
OHL = np.dot(HLW,X1)
OHL
```

```python
#Output of the final layer
OOL = np.dot(OLW, OHL)
OOL
```

2. What is the network output of the input (1,2)?

```python
X2 = [1,2]
X2 = np.array(X2)
```

```python
#Output of the hidden layer
OHL2 = np.dot(HLW,X2)
OHL2
```

```python
#Output of the final layer
OOL2 = np.dot(OLW, OHL2)
OOL2
```

Do you do all the network computations once again to answer this question. Explain why you do or do not have to do this

R: As the values are double the initial values and the function is linear with a bias vector of 0, the output of the final layer is also doubled. Therefore, there is no need to compute the result again; it can simply be scaled by a factor of 2.
